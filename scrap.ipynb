#### This codes scraps HMI Intensitygrams (Flattened, orange) in 1024x1024 format from the NASA website for the whole month of January 2025, and puts them in the images folder. 

Total execution takes ~ 40min, with 1623 images taking 500Mo

```python
import requests
from bs4 import BeautifulSoup

hyperlinks = []
for day in range(1,31): #You can dowload less by replacing 31 with a smaller number
    url = f"https://sdo.gsfc.nasa.gov/assets/img/browse/2025/01/{'0' if day<10 else ''}{day}/"
    #get the content
    response = requests.get(url)

    # Parse the HTML content of the page
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all hyperlinks on the page
    links = soup.find_all('a')

    # Extract the href attribute from each hyperlink that matches with HMI intensitygram flattened
    hyperlinks += [url+link.get('href') for link in links
                    if link.get('href').endswith('1024_HMIIF.jpg')
                    ]


hyperlinks
```

```python
import os
from pathlib import Path
import time

# Create images directory if it doesn't exist
Path('images').mkdir(exist_ok=True)

# Download each image
for link in hyperlinks:
    # Extract filename from URL
    filename = link.split('/')[-1]
    save_path = os.path.join('images', filename)
    
    # Download the image
    response = requests.get(link)
    
    # Save the image if request was successful
    if response.status_code == 200:
        with open(save_path, 'wb') as f:
            f.write(response.content)
    
    # Add a small delay to be respectful to the server
    time.sleep(0.1)

print("Download completed!")
```
